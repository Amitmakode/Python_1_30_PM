```python
# Let's start by loading the diabetes dataset from sklearn, then convert it to a pandas DataFrame for exploratory data analysis (EDA).

import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt
import seaborn as sns

# Load diabetes dataset
diabetes = load_diabetes()
```

```python
diabetes
```

```python
# Convert to DataFrame
df = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target  # Add target variable to the dataframe
```

```python
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe().T
```

```python
X = df.drop("target", axis = 1)
```

```python
X
```

```python
y = df.target
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state= 45)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
```

```python
x_train_scale= scale.fit_transform(x_train)
```

```python
x_test_scale = scale.transform(x_test)
```

```python
x_train_scale
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr =lr.fit(x_train_scale, y_train)
```

```python
lr.score(x_test_scale,y_test)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
y_predict = lr.predict(x_test_scale)
```

```python
y_predict.shape
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
MSE = mean_squared_error(y_test, y_predict)
```

```python
MAE = mean_absolute_error(y_test, y_predict)
```

```python
RMSE = np.sqrt(MSE)
```

```python
R2 = r2_score(y_test,y_predict)
```

```python
R2
```
